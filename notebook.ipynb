## Exploratory Data Analysis (EDA)

Melakukan eksplorasi terhadapat data day untuk mempelajari apa yang ada dalam data day.csv

````python
day_df.groupby(by="mnth").agg({
      "instant": "nunique",
      "mnth": ["max", "min", "mean", "std"]
  })
````

Fungsi kode diatas untuk memahami data bulan pada day.csv

````python
# Menghitung rata-rata rental bike per month
month_counts = day_df.groupby('mnth')['cnt'].mean()

print(month_counts)
````

## Visualization & Explanatory Analysis

### Pertanyaan 1:
Bagaimana performa rental bike dalam 1 bulan ?

````python
# Plotting a pie chart
plt.pie(month_counts, labels=[
    'Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli',
    'Agustus', 'September', 'Oktober', 'November', 'Desember'
], autopct='%1.1f%%', startangle=90)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Rata-Rata Dalam Sebulan')
plt.show()
````

### Pertanyaan 2:
Jam berapa sepeda sering dan jarang disewa ?

````python
# Calculate total rentals per hour
total_rentals_per_hour = hour_df.groupby(hour_df['hr'])['cnt'].sum()

# Plotting a line chart
plt.plot(total_rentals_per_hour.index, total_rentals_per_hour.values, marker='o')
plt.xlabel('Jam')
plt.ylabel('Jumlah Rental Bike')
plt.title('Total Rentals Per Hour')
plt.show()
````

````python
# Combine DataFrames vertically (along rows)
combined_df = pd.concat([day_df, hour_df], axis=0)

# Save the combined DataFrame to a CSV file
combined_df.to_csv("main_data.csv", index=False)
````
